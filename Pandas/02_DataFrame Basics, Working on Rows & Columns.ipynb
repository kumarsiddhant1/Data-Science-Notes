```python
print(df['total_bill']) # Accessing Single column. Will be a Series
type(df['total_bill'])
```

```python
# We can grab a row in 2 ways: 
# 1.) Using default index
# 2.) Using labeled index
```

```python
# Using default index
# We use iloc for for this.

print(type(df.iloc[0]))

df.iloc[0]
```

```python
df.iloc[0:4] # Grabbing multiple rows using iloc.
```

```python
# Using labeled index.
# we use it if we know the label.
# We use loc for for this.

print(type(df.loc["Sun4458"]))

df.loc["Sun4458"]
```

```python
df.loc[["Sun4458", "Sun5260"]]  # Grabbing multiple rows using loc.
```

```python
row = df.loc['Sun4458'] #for later use
row
```

```python
#Dropping a Row
df.head()
df = df.drop('Sun4458',axis=0)
df.head()
```

```python
df
```
